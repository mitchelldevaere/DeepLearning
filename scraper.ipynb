```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

# Global counter for all downloaded images
global_counter = 1

# Function to download an image
def download_image(url, folder):
    global global_counter

    try:
        # Extract the filename from the URL
        filename = os.path.join(folder, f"{global_counter:03d}")

        # Check if the file extension is '.gif' and skip download if true
        if filename.lower().endswith('.gif'):
            print(f"Skipped downloading GIF: {url}")
            return 0

        response = requests.get(url, stream=True)
        if response.status_code == 200:
            # Save the image
            with open(filename, 'wb') as file:
                for chunk in response.iter_content(chunk_size=128):
                    file.write(chunk)

            print(f"Downloaded: {filename}")
            global_counter += 1
            return 1
        else:
            print(f"Failed to download: {url}")
            return 0
    except Exception as e:
        print(f"Error downloading {url}: {str(e)}")
        return 0

# Function to scrape images from a page
def scrape_images(url, folder):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find the table containing images
            table = soup.find('table')
            if table:
                # Extract image URLs from the table
                image_urls = [urljoin(url, img['src']) for img in table.find_all('img')]

                # Download each image and count the downloaded images
                total_downloaded = 0
                for img_url in image_urls:
                    total_downloaded += download_image(img_url, folder)

                print(f"Images from {url}: {total_downloaded}")
                return total_downloaded
            else:
                print(f"No table found on the page: {url}")
                return 0
        else:
            print(f"Failed to fetch page: {url}")
            return 0
    except Exception as e:
        print(f"Error scraping {url}: {str(e)}")
        return 0


urls = [
    "http://www.rembrandtpainting.net/complete_catalogue/start_history.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_history_two.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_landscapes.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_self_portraits.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_portraits.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_family_portraits.htm",
    "http://www.rembrandtpainting.net/complete_catalogue/start_allegories.htm"
]


# Main function
def main():
    # Create a folder to store the downloaded images
    folder = 'downloaded_images'
    os.makedirs(folder, exist_ok=True)

    # Initialize counters
    global global_counter
    global_counter = 1

    total_images = 0

    # Loop through the provided URLs
    for page_url in urls:
        images_from_url = scrape_images(page_url, folder)
        total_images += images_from_url

    print(f"\nTotal images downloaded: {total_images}")

if __name__ == "__main__":
    main()
```
